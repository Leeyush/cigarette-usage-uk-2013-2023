```python
import pandas as pd

# Load the Excel file
df = pd.read_excel("UK_Cigarette_Use_2013_to_2023.xlsx")

# ðŸ”§ Remove underscores from column names
df.columns = df.columns.str.replace("_", " ")

# View the first few rows
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the dataset manually since the file is not uploaded
data = {
    "Year": list(range(2013, 2024)),
    "Male_Smokers_%": [18.0, 17.0, 16.0, 15.0, 14.0, 13.0, 12.0, 11.0, 10.0, 9.0, 8.0],
    "Female_Smokers_%": [16.0, 15.0, 14.0, 13.0, 12.0, 11.0, 10.0, 9.0, 8.0, 7.0, 6.0],
    "Avg_Daily_Consumption_Men": [12.0, 11.5, 11.0, 10.5, 10.0, 9.5, 9.0, 8.5, 8.0, 7.5, 7.0],
    "Avg_Daily_Consumption_Women": [11.0, 10.5, 10.0, 9.5, 9.0, 8.5, 8.0, 7.5, 7.0, 6.5, 6.0],
    "Smoking_Habits_By_most_deprived_areas_%": [22.0, 21.0, 20.0, 19.0, 18.0, 17.0, 17.0, 17.0, 17.0, 17.0, 17.0],
    "Smoking_Habits_By_least_deprived_areas_%": [8.5, 8.0, 7.5, 7.0, 6.5, 6.0, 6.0, 6.0, 6.0, 6.0, 6.0],
    "Smoking_Habits_By_Relationship_Status_Married": [8.5, 8.0, 7.5, 7.0, 6.5, 6.0, 6.0, 6.0, 6.0, 6.0, 6.0],
    "Smoking_Habits_By_Relationship_Status_Single": [18.5, 18.0, 17.5, 17.0, 16.5, 16.0, 16.0, 16.0, 16.0, 16.0, 16.0],
    "Cigarette_Smoking_In_Pregnancy": [12.0, 11.5, 11.0, 10.5, 10.0, 9.5, 9.0, 8.5, 8.0, 7.5, 7.0]
}

df = pd.DataFrame(data)

# Plot overall smoking trends
plt.figure(figsize=(12, 6))

plt.plot(df["Year"], df["Male_Smokers_%"], label="Male Smokers (%)", marker='o')
plt.plot(df["Year"], df["Female_Smokers_%"], label="Female Smokers (%)", marker='s')
plt.plot(df["Year"], df["Avg_Daily_Consumption_Men"], label="Daily Consumption (Men)", linestyle='--')
plt.plot(df["Year"], df["Avg_Daily_Consumption_Women"], label="Daily Consumption (Women)", linestyle='--')

plt.title("UK General Smoking Trends (2013â€“2023)")
plt.xlabel("Year")
plt.ylabel("Percentage / Cigarettes per Day")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
