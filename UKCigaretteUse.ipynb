```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Define X and y
X = df["Year"].values.reshape(-1, 1)
y = df["Avg_Daily_Consumption_Men"].values

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict next 5 years
future_years = np.array(range(2024, 2029)).reshape(-1, 1)
predictions = model.predict(future_years)

# Plot
plt.scatter(X, y, label="Actual")
plt.plot(future_years, predictions, label="Forecast", color="red")
plt.title("Forecast: Avg Daily Consumption (Men)")
plt.xlabel("Year")
plt.ylabel("Cigarettes per Day")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Output forecast
for year, value in zip(future_years.flatten(), predictions):
    print(f"{year}: {value:.2f} cigarettes/day")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Make sure your 'Year' column is datetime
df["Year"] = pd.to_datetime(df["Year"], format="%Y")
df.set_index("Year", inplace=True)
```

```python
# Linear Regression for Female Smokers %
X = df.index.year.values.reshape(-1, 1)
y = df["Female_Smokers_%"].values

model_lr = LinearRegression()
model_lr.fit(X, y)

# Forecast from 2024 to 2030
future_years = np.arange(2024, 2031).reshape(-1, 1)
predictions_lr = model_lr.predict(future_years)

# Plot
plt.figure(figsize=(10,6))
plt.plot(X, y, label="Actual", marker='o')
plt.plot(future_years, predictions_lr, label="Linear Forecast", color='red', marker='o')
plt.title("Linear Regression: Female Smokers %")
plt.xlabel("Year")
plt.ylabel("%")
plt.legend()
plt.grid(True)
plt.show()

# Print Forecast
for yr, pred in zip(future_years.ravel(), predictions_lr):
    print(f"{yr}: {pred:.2f}%")
```

```python
series_female = df["Female_Smokers_%"]

# Fit ARIMA
model_arima = ARIMA(series_female, order=(1,1,1))
model_fit_arima = model_arima.fit()

# Forecast to 2030
forecast_obj = model_fit_arima.get_forecast(steps=7)
forecast_mean = forecast_obj.predicted_mean
conf_int = forecast_obj.conf_int()

# Create future index
future_index = pd.date_range(start="2024", periods=7, freq="Y")
forecast_mean.index = future_index
conf_int.index = future_index

# Plot
plt.figure(figsize=(10,6))
plt.plot(series_female.index, series_female, label="Actual", marker='o')
plt.plot(forecast_mean.index, forecast_mean, label="ARIMA Forecast", color='green', marker='o')
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='lightgreen', alpha=0.4)
plt.title("ARIMA Forecast: Female Smokers %")
plt.xlabel("Year")
plt.ylabel("%")
plt.grid(True)
plt.legend()
plt.show()
```
